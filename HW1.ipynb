```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Left, right and midpoint Riemann sums for the function f(x)= x*x

over the interval [4,10]with a partition of size N=20

```python
f = lambda x : x*x
a = 4; b = 10; N = 20
n = 20 # Use n*N+1 points to plot the function smoothly

x = np.linspace(a,b,N+1)
y = f(x)

X = np.linspace(a,b,n*N+1)
Y = f(X)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.plot(X,Y,'b')
xleft = x[:-1] # Left endpoints
yleft = y[:-1]
plt.plot(xleft,yleft,'b.',markersize=10)
plt.bar(xleft,yleft,width=(b-a)/N,alpha=0.2,align='edge',edgecolor='b')
plt.title('Left Riemann Sum, N = {}'.format(N))

plt.subplot(1,3,2)
plt.plot(X,Y,'b')
xmid = (x[:-1] + x[1:])/2 # Midpoints
ymid = f(xmid)
plt.plot(xmid,ymid,'b.',markersize=10)
plt.bar(xmid,ymid,width=(b-a)/N,alpha=0.2,edgecolor='b')
plt.title('Midpoint Riemann Sum, N = {}'.format(N))

plt.subplot(1,3,3)
plt.plot(X,Y,'b')
xright = x[1:] # Right endpoints
yright = y[1:]
plt.plot(xright,yright,'b.',markersize=10)
plt.bar(xright,yright,width=-(b-a)/N,alpha=0.2,align='edge',edgecolor='b')
plt.title('Right Riemann Sum, N = {}'.format(N))

plt.show()
```

```python
dx = (b-a)/N
xleft = np.linspace(a,b-dx,N)
xmidpoint = np.linspace(dx/2,b - dx/2,N)
xright = np.linspace(dx,b,N)

print("Partition with",N,"subintervals.")
leftriemannsum = np.sum(f(xleft) * dx)
print("Left Riemann Sum:",leftriemannsum)

midpointriemannsum = np.sum(f(xmidpoint) * dx)
print("Midpoint Riemann Sum:",midpointriemannsum)

rightriemannsum = np.sum(f(xright) * dx)
print("Right Riemann Sum:",rightriemannsum)
```

# Question 2
We want to sample random samples from the unit circle 
. Write a python function which takes one parameter  the number of sample points and returns a sample  points on the unit circle as a  numpy matrix. Calculate some examples, and visualize them.

```python
import random
import math
circlex = 200
circley = 200
a = random.randint(0,200) * 2 * math.pi
r = 1 * math.sqrt(random.randint(0,200))
x = r * math.cos(a) + circlex
y = r * math.sin(a) + circley
```

```python
randProba = lambda a: a/sum(a)
npoints = 1000 # points to chose from
r = 3 # radius of the circle

plt.figure(figsize=(4,4))
t = np.linspace(0, 2*np.pi, npoints, endpoint=False)
x = r * np.cos(t)
y = r * np.sin(t)
plt.scatter(x, y, c='0.8')

n = 1 # number of points to chose
t = np.linspace(0, 2*np.pi, npoints, endpoint=False)[np.random.choice(range(npoints), n, replace=False, 
                                                                      p=randProba(np.random.random(npoints)))]
x = r * np.cos(t)
y = r * np.sin(t)

plt.scatter(x, y)
```
